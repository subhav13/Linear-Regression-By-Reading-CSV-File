```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds=pd.read_csv('Position_Salaries.csv')
ds.head()
```

```python
x=ds.iloc[:,1:2].values
y=ds.iloc[:,2].values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg= PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x)
```

```python
lin_reg2=LinearRegression()
lin_reg2.fit(x_poly,y)
```

```python
#linear prediction 
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='black')
plt.title('TRUTH VS BLUFF(LINEAR PREDICTION)')
plt.xlabel('LEVELS')
plt.ylabel('SALARIES')
plt.show()

#Polynomial prediction
plt.scatter(x,y,color='green')
plt.plot(x,lin_reg2.predict(poly_reg.fit_transform(x)),color='black')
plt.title('TRUTH VS BLUFF(POLYNOMIAL PREDICTION)')
plt.xlabel('LEVELS')
plt.ylabel('SALARIES')
plt.show()

# A more accurate polynomial prediction
x_grid=np.arange(min(x),max(x),0.1)
x_grid=x_grid.reshape((len(x_grid),1))
plt.scatter(x,y,color='blue')
plt.plot(x_grid,lin_reg2.predict(poly_reg.fit_transform(x_grid)),color='black')
plt.title('TRUTH VS BLUFF(A MORE ACCURATE POLYNOMIAL PREDICTION)')
plt.xlabel('LEVELS')
plt.ylabel('SALARIES')
plt.show()
```
